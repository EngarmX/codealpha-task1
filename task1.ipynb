```python
X = data.drop('label', axis=1)  # Features
y = data['label']  # Target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
num_folds = 5
model = LogisticRegression()
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)
cross_val_predictions = cross_val_predict(model, X, y, cv=kf)
```

```python
accuracy_cv = accuracy_score(y, cross_val_predictions)
conf_matrix_cv = confusion_matrix(y, cross_val_predictions)
class_report_cv = classification_report(y, cross_val_predictions)
```

```python
print(accuracy_cv)
print(conf_matrix_cv)
print(class_report_cv)
```

```python
result_df = pd.DataFrame({'Actual': y, 'Predicted': cross_val_predictions})

# Display the DataFrame for the entire dataset
print("Results on Entire Dataset:")
print(result_df.head(50))
```
